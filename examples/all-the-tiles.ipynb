```python
%matplotlib inline
import matplotlib.pyplot as pyplot

import sys
sys.path.insert(0, "../")
from weavingspace import TileShape
from weavingspace import TileUnit
```

# All the tiles!
## Basic tile units

```python
fig = pyplot.figure(figsize = (24, 8))
units = [TileUnit(), 
         TileUnit(base_shape = TileShape.HEXAGON),
         TileUnit(base_shape = TileShape.TRIANGLE)]
for u, posn in zip(units, range(1, 4)):
    ax = fig.add_subplot(1, 3, posn)
    u.plot(ax = ax, r = 1, show_vectors = True, show_reg_prototile = False)
    pyplot.axis("off")
```

## Cairo
The one true tiling.

```python
t = TileUnit(tiling_type = "cairo")
fig = pyplot.figure(figsize = (8, 8))
ax = fig.add_subplot(111)
fig = t.plot(ax = ax, r = 1, show_vectors = True, show_reg_prototile = False)
ax.set_axis_off()
```

## Hexagon colourings
Only 3, 4, 7 are supported for now...

```python
units = [
    TileUnit(tiling_type = "hex-colouring", n = 3),
    TileUnit(tiling_type = "hex-colouring", n = 4),    
    TileUnit(tiling_type = "hex-colouring", n = 7)]
fig = pyplot.figure(figsize = (24, 8))
for u, posn in zip(units, range(1, 4)):
    ax = fig.add_subplot(1, 3, posn)
    ax.set_title(f"n={u.n}")
    u.plot(ax = ax, r = 1, show_vectors = True, show_reg_prototile = False)
    pyplot.axis("off")    
```

## A square colouring
Just the one for now, because it's nice to have a 5-colour option.

```python
units = [
    TileUnit(tiling_type = "square-colouring", n = 5)]
fig = pyplot.figure(figsize = (7, 7))
for u, posn in zip(units, range(1, 2)):
    ax = fig.add_subplot(1, 1, posn)
    ax.set_title(f"n={u.n}")
    u.plot(ax = ax, r = 1, show_vectors = True, show_reg_prototile = False)
    pyplot.axis("off")    
```

## Hexagon slices

```python
units = [
    TileUnit(tiling_type = "hex-slice", n = 2, offset = 0), 
    TileUnit(tiling_type = "hex-slice", n = 2, offset = 1), 
    TileUnit(tiling_type = "hex-slice", n = 3, offset = 0),
    TileUnit(tiling_type = "hex-slice", n = 3, offset = 1), 
    TileUnit(tiling_type = "hex-slice", n = 4, offset = 0),
    TileUnit(tiling_type = "hex-slice", n = 4, offset = 1),
    TileUnit(tiling_type = "hex-slice", n = 6, offset = 0), 
    TileUnit(tiling_type = "hex-slice", n = 6, offset = 1),
    TileUnit(tiling_type = "hex-slice", n = 12, offset = 0),    
    TileUnit(tiling_type = "hex-slice", n = 12, offset = 1)]
fig = pyplot.figure(figsize = (24, 32))
for u, posn in zip(units, range(1, 11)):
    ax = fig.add_subplot(4, 3, posn)
    ax.set_title(f"n={u.n}, dissection_offset={u.offset}")
    u.plot(ax = ax, r = 1, show_vectors = True, show_reg_prototile = False)
    pyplot.axis("off")
```

## Hexagon dissections

```python
units = [
    TileUnit(tiling_type = "hex-dissection", n = 4, offset = 0), TileUnit(tiling_type = "hex-dissection", n = 4, offset = 1),
    TileUnit(tiling_type = "hex-dissection", n = 7, offset = 0), TileUnit(tiling_type = "hex-dissection", n = 7, offset = 1),
    TileUnit(tiling_type = "hex-dissection", n = 9, offset = 0), TileUnit(tiling_type = "hex-dissection", n = 9, offset = 1)]
fig = pyplot.figure(figsize = (18, 12))
for u, posn in zip(units, range(1, 7)):
    ax = fig.add_subplot(2, 3, posn)
    ax.set_title(f"n={u.n}, dissection_offset={u.offset}")
    u.plot(ax = ax, r = 1, show_vectors = True, show_reg_prototile = False, cmap = "tab10")
    pyplot.axis("off")
```

## Laves tile units
Some of these are determined using the hex dissection.

```python
units = [TileUnit(tiling_type = "laves", code = "3.3.3.3.6"), 
         TileUnit(tiling_type = "laves", code = "3.3.4.3.4"),
         TileUnit(tiling_type = "laves", code = "3.4.6.4"), 
         TileUnit(tiling_type = "laves", code = "3.6.3.6"),
         TileUnit(tiling_type = "laves", code = "3.12.12"),
         TileUnit(tiling_type = "laves", code = "4.6.12"),
         TileUnit(tiling_type = "laves", code = "4.8.8")]
fig = pyplot.figure(figsize = (24, 24))
for u, posn in zip(units, range(1, 8)):
    ax = fig.add_subplot(3, 3, posn)
    ax.set_title(f"({u.code})")
    u.plot(ax = ax, r = 1, show_vectors = True, show_reg_prototile = False)
    pyplot.axis("off")
```
