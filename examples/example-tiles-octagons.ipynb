```python
%matplotlib inline

import geopandas as gpd

import sys
sys.path.insert(0, "../")
from weavingspace.tile_unit import TileUnit
from weavingspace.tile_map import Tiling
```

```python
ak = gpd.read_file("data/imd-auckland-2018.gpkg")
```

# Octagons (4.8.8)

```python
d = 400
tile_unit = TileUnit(tiling_type = "archimedean", code = "4.8.8",
                     spacing = d, crs = ak.crs)
tile_unit.plot(r = 1)
```

```python
tiling = Tiling(tile_unit, ak, tiles_margin = 10)
```

```python
t = tiling.get_tiled_map()
```

```python
t.variables = dict(zip(list("ab"), [x for x in ak if "Rank" in x]))
t.colourmaps = dict(zip(t.variables.values(), ["Oranges", "Blues"]))

fig = t.render(legend_zoom = 0.9, use_ellipse = True, 
               ellipse_magnification = 1.2)
```
