```python
%matplotlib inline

import geopandas as gpd

import sys
sys.path.insert(0, "../")
from weavingspace import TileUnit
from weavingspace import Tiling
```

```python
ak = gpd.read_file("data/imd-auckland-2018.gpkg")
```

# Octagons (4.8.8)

```python
d = 400
tile_unit = TileUnit(tiling_type = "archimedean", code = "4.8.8",
                     spacing = d, crs = ak.crs)
tile_unit.plot(r = 1)
```

```python
tiling = Tiling(tile_unit, ak, tiles_margin = 10)
```

```python
t = tiling.get_tiled_map()
```

```python
t.ids_to_map = list("ab")
t.vars_to_map = [x for x in ak if "Rank" in x]
t.colors_to_use = "Reds"
t.n_classes = 5

fig = t.render(legend_zoom = 0.9, use_ellipse = True, 
               ellipse_magnification = 1.2)
```
