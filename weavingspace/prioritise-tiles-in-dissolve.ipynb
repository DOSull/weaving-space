```python
%matplotlib inline
from matplotlib import pyplot
import geopandas

from weave_units import WeaveUnit
from tile_map import Tiling


ak = geopandas.read_file("../data/imd-auckland-2018.gpkg")
unit = WeaveUnit(spacing = 200, crs = ak.crs, aspect = 0.8, 
                 strands = "abc|defg")
tiling = Tiling(unit, ak, id_var = "DZ2018")
```

```python
t30a = tiling.get_tiled_map(rotation = 30, prioritise_tiles = True)
t30b = tiling.get_tiled_map(rotation = 30, prioritise_tiles = False)
```

```python
fig = pyplot.figure(figsize = (24, 15))
ax = fig.add_subplot(111)

strands = set(t30a.element_id)
cols = [x for x in t30a.columns if "Rank" in x]
cmaps = ("OrRd", "BuPu", "Greys", "BuGn", "YlOrRd", "YlGnBu", "RdPu")

for strand, col, cmap in zip(strands, cols, cmaps):
    t30a[t30a.element_id == strand].plot(ax = ax, column = col, cmap = cmap, k = 100, scheme = "equal_interval", linewidth = 0)
```

```python
fig = pyplot.figure(figsize = (24, 15))
ax = fig.add_subplot(111)

strands = set(t30b.element_id)
cols = [x for x in t30b.columns if "Rank" in x]
cmaps = ("OrRd", "BuPu", "Greys", "BuGn", "YlOrRd", "YlGnBu", "RdPu")

for strand, col, cmap in zip(strands, cols, cmaps):
    t30b[t30b.element_id == strand].plot(ax = ax, column = col, cmap = cmap, k = 100, scheme = "equal_interval", linewidth = 0)
```
