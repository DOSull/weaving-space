# Hexagon dissections
We can dissect a hexagon radially into 2, 3, 4, 6 or 12 congruent tiles. We can start the dissections at a hexagon corner (`dissection_offset == 0`) or edge midpoint (`dissection_offset == 1`).

The code to make the tile unit looks like this:

```python
n = 2
tile_unit_1 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 0)
tile_unit_2 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 1)
```

```python
fig = pyplot.figure(figsize = (16, 8))
ax = fig.add_subplot(121)
tile_unit_1.plot(ax = ax, r = 1)

ax = fig.add_subplot(122)
tile_unit_2.plot(ax = ax, r = 1)
```

## 2-dissections
Here `dissection_offset = 1` is very similar to Laves \[3<sup>3</sup>.4<sup>2</sup>\].

```python
display(fig)
```

```python
n = 3
tile_unit_1 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 0)
tile_unit_2 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 1)

fig = pyplot.figure(figsize = (16, 8))
ax = fig.add_subplot(121)
tile_unit_1.plot(ax = ax, r = 1)

ax = fig.add_subplot(122)
tile_unit_2.plot(ax = ax, r = 1)
```

## 3-dissections
Here `offset_option = 0` is equivalent to 'madweave' in 3 colours, or to Laves \[3.6.3.6\]

```python
display(fig)
```

```python
n = 4
tile_unit_1 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 0)
tile_unit_2 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 1)

fig = pyplot.figure(figsize = (16, 8))
ax = fig.add_subplot(121)
tile_unit_1.plot(ax = ax, r = 1)

ax = fig.add_subplot(122)
tile_unit_2.plot(ax = ax, r = 1)
```

## 4-dissections
The two options here are equivalent under rotation.

```python
display(fig)
```

```python
n = 6
tile_unit_1 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 0)
tile_unit_2 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 1)

fig = pyplot.figure(figsize = (16, 8))
ax = fig.add_subplot(121)
tile_unit_1.plot(ax = ax, r = 1)

ax = fig.add_subplot(122)
tile_unit_2.plot(ax = ax, r = 1)
```

## 6-dissections
`dissection_offset = 0` is equivalent to \(3<sup>6</sup>\) and Laves \[6<sup>3</sup>\]. Offset 1 is equivalent to Laves \[3.4.6.4\]

```python
display(fig)
```

```python
n = 12
tile_unit_1 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 0)
tile_unit_2 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 1)

fig = pyplot.figure(figsize = (16, 8))
ax = fig.add_subplot(121)
tile_unit_1.plot(ax = ax, r = 1)

ax = fig.add_subplot(122)
tile_unit_2.plot(ax = ax, r = 1)
```
