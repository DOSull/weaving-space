# Hexagon dissections
We can dissect a hexagon radially into 2, 3, 4, 6 or 12 congruent tiles. We can start the dissections at a hexagon corner (`dissection_offset == 0`) or edge midpoint (`dissection_offset == 1`).

The code to make the tile unit looks like this:

```python
n = 2
tile_unit_1 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 0)
tile_unit_2 = TileUnit(
    tiling_type = "hex-dissection", n = n, dissection_offset = 1)
```

```python
fig = pyplot.figure(figsize = (16, 8))
ax = fig.add_subplot(121)
tile_unit_1.get_local_patch(r = 1).plot(ax = ax, column = "element_id", alpha = 0.3, cmap = "Paired")
tile_unit_1.elements.plot(ax = ax, column = "element_id", cmap = "Paired")
tile_unit_1.tile.plot(ax = ax, edgecolor = "k", facecolor = "#00000000")
 
ax = fig.add_subplot(122)
tile_unit_2.get_local_patch(r = 1).plot(ax = ax, column = "element_id", alpha = 0.3, cmap = "Paired")
tile_unit_2.elements.plot(ax = ax, column = "element_id", cmap = "Paired")
tile_unit_2.tile.plot(ax = ax, edgecolor = "k", facecolor = "#00000000")
```
