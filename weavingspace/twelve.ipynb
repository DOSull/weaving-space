```python
sf = gpd.read_file("../data/sf_demo.gpkg").to_crs(2225)
```

## Legend and map utils

```python
tile_unit = TileUnit(tiling_type = "archimedean", code = "3.3.4.3.4", 
                     spacing = 1500, crs = sf.crs, margin = 0.0)
tile_unit.plot(r = 1)
```

```python
tiling = Tiling(tile_unit, sf, id_var = "id")
```

```python
t30 = tiling.get_tiled_map(prioritise_tiles = True, rotation = 30)
```

```python
columns = list(t30.columns)[7:19]
cmaps = ("Reds", "Blues", "Greens", "Oranges", "Purples", "Greys") * 2
cmaps = ("OrRd", "YlGnBu", "YlGn", "YlOrRd", "BuPu", "RdPu") * 2

fig = pyplot.figure(figsize = (25, 15))
tiling.plot_map(fig, t30, vars = columns, pals = cmaps, 
                rotate_text = False, legend = True)
```
